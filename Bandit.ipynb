```python
import numpy as np

class Bandit():
    
    def __init__(self, N, est=None, counts=None):
        self.Nmachines=N
        self.arm_values=np.random.rand(N)
        
        if est is None:
            self.est_values = np.zeros(N)
        else:
            self.est_values = est
        
        if counts is None:
            self.useage_values = np.zeros(N)
        else:
            self.useage_values = counts
    
    def get_reward(self,action):
        rand_num=np.random.random()
        if rand_num < self.arm_values[action]:
            return 1
        else:
            return 0
    
    def greedy_hour_select(self, epsilon):
        rand_num=np.random.random()
        if rand_num>epsilon:
            return np.argmax(self.est_values, axis=0)
        else:
            a=np.random.randint(self.Nmachines)
            return (a)
          
    def UCP_hour_select(self):
        totalN = self.useage_values.sum(axis=1).reshape(-1, 1)
        uncertanty = np.sqrt(2*np.log(totalN+1)/self.useage_values)
        return np.argmax(self.est_values + uncertanty, axis=1)

    def update_est(self, action, reward):
        for row, index in enumerate(action):
            self.useage_values[row, index]+=1
            self.est_values[row,index] = (self.est_values[row, index]*(self.useage_values[row, index]-1)+reward[row])/np.float(self.useage_values[row, index])

    def bandit_test(self, NPull, epsilon=0.02):
        reward_history=[]
        for i in range(NPull):
            # machine=bandit.greedy_machine_select(epsilon)
            machine = bandit.UCP_hour_select()
            reward=bandit.get_reward(machine)
            bandit.update_est(machine,reward)
            reward_history.append(reward)
        print(self.est_values)
        print(self.useage_values)
```
